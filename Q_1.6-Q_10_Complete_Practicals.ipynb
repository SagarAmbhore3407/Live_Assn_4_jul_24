```python
# 1.7. Write a program to create a simple calculator performing only four basic operations(+,-,/,*) .

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error! Division by zero."

def calculator():
    while True:
        try:
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
            break
        except ValueError:
            print("Invalid input. Please enter numeric values.")

    while True:
        operator = input("Enter the operation (+, -, *, /): ").strip()
        if operator in ('+', '-', '*', '/'):
            break
        else:
            print("Invalid operator. Please enter one of +, -, *, /.")

    if operator == '+':
        print(f"{num1} + {num2} = {add(num1, num2)}")
    elif operator == '-':
        print(f"{num1} - {num2} = {subtract(num1, num2)}")
    elif operator == '*':
        print(f"{num1} * {num2} = {multiply(num1, num2)}")
    elif operator == '/':
        result = divide(num1, num2)
        print(f"{num1} / {num2} = {result}")

calculator()
```

```python
#  1.9.  Write a program to find the factors of a whole number  using a while loop.

def find_factors(number):
    factors = []
    i = 1
    while i <= number:
        if number % i == 0:
            factors.append(i)
        i += 1
    return factors

# Input: a whole number from the user
while True:
    try:
        number = int(input("Enter a whole number: "))
        if number < 0:
            raise ValueError("The number must be a non-negative whole number.")
        break
    except ValueError as e:
        print(f"Invalid input. Please enter a valid whole number. ({e})")

# Find and print the factors
factors = find_factors(number)
print(f"The factors of {number} are: {factors}")
```

```python
# 1.10. Write a program to find the sum of all the positive numbers entered by the user. As soon as the user 
#       enters a negative number, stop taking in any further input from the user and display the sum .

def sum_of_positive_numbers():
    total_sum = 0
    
    while True:
        try:
            number = float(input("Enter a positive number (or a negative number to stop): "))
            
            if number < 0:
                break  # Exit the loop if the number is negative
            
            if number > 0:
                total_sum += number
            else:
                print("Please enter a positive number or a negative number to stop.")
        
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
    
    return total_sum

# Call the function and display the result
result = sum_of_positive_numbers()
print(f"The sum of all positive numbers entered is: {result}")
```

```python
#  1.11.   Write a program to find prime numbers between 2 to 100 using nested for loops.

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_primes(start, end):
    prime_numbers = []
    for num in range(start, end + 1):
        if is_prime(num):
            prime_numbers.append(num)
    return prime_numbers

# Find and print prime numbers between 2 and 100
primes = find_primes(2, 100)
print(f"Prime numbers between 2 and 100 are: {primes}")
```

```python
#  1.12.   Write the  programs for the following
#          Accept the marks of the student in five major subjects and display the same
#          Calculate the sum of the marks of all subjects.Divide the total marks by number of subjects (i.e. 5),
#          calculate percentage = total marks/5 and display the percentage
#          Find the grade of the student as per the following criteria . Hint: Use Match & case for this.:

'''percentage > 85   --- A
percentage < 85 && percentage >= 75 ----B
percentage < 75 && percentage >= 50 ----C
percentage >30 && percentage <=  50 ----D
percentage  < 30   --- Reappear'''


def calculate_grade(percentage):
    match percentage:
        case _ if percentage > 85:
            return 'A'
        case _ if 75 <= percentage <= 85:
            return 'B'
        case _ if 50 <= percentage < 75:
            return 'C'
        case _ if 30 <= percentage < 50:
            return 'D'
        case _ if percentage < 30:
            return 'Reappear'
        case _:
            return 'Invalid Percentage'

def main():
    # Input marks for five subjects
    marks = []
    subjects = 5
    for i in range(subjects):
        while True:
            try:
                mark = float(input(f"Enter the marks for subject {i + 1}: "))
                if mark < 0:
                    raise ValueError("Marks cannot be negative. Please enter a valid mark.")
                marks.append(mark)
                break
            except ValueError as e:
                print(f"Invalid input. {e}")
    
    # Display the marks
    print("\nMarks in each subject:")
    for i, mark in enumerate(marks, start=1):
        print(f"Subject {i}: {mark}")
    
    # Calculate total and percentage
    total_marks = sum(marks)
    percentage = total_marks / subjects

    print(f"\nTotal Marks: {total_marks}")
    print(f"Percentage: {percentage:.2f}%")
    
    # Determine and display grade
    grade = calculate_grade(percentage)
    print(f"Grade: {grade}")

# Run the program
main()

```

```python

#1.14. Consider the gravitational interactions between the Earth, Moon, and Sun in our solar system. 
# Given:
# mass_earth = 5.972e24  # Mass of Earth in kilograms
# mass_moon = 7.34767309e22  # Mass of Moon in kilograms
# mass_sun = 1.989e30  # Mass of Sun in kilograms
# distance_earth_sun = 1.496e11  # Average distance between Earth and Sun in meters
# distance_moon_earth = 3.844e8  # Average distance between Moon and Earth in meters
# Tasks: 
# Calculate the gravitational force between the Earth and the Sun
# Calculate the gravitational force between the Moon and the Earth
# Compare the calculated forces to determine which gravitational force is stronger
# Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison.





def calculate_gravitational_force(m1, m2, distance):
    
    '''Uses Newton's law of universal gravitation to compute the force between two bodies given their masses and distance.'''

    G = 6.67430e-11  
    return G * m1 * m2 / (distance ** 2)



# Given data
mass_earth = 5.972e24  # Mass of Earth in kg
mass_moon = 7.34767309e22  # Mass of Moon in kg
mass_sun = 1.989e30  # Mass of Sun in kg
distance_earth_sun = 1.496e11  # Distance between Earth and Sun in meters
distance_moon_earth = 3.844e8  # Distance between Moon and Earth in meters

# Calculate gravitational forces
force_earth_sun = calculate_gravitational_force(mass_earth, mass_sun, distance_earth_sun)
force_moon_earth = calculate_gravitational_force(mass_moon, mass_earth, distance_moon_earth)

# Compare the forces
if force_earth_sun > force_moon_earth:
    stronger_force = "Earth-Sun"
else:
    stronger_force = "Moon-Earth"

# Print results
print(f"Gravitational force between Earth and Sun: {force_earth_sun:.2e} N")
print(f"Gravitational force between Moon and Earth: {force_moon_earth:.2e} N")
print(f"The stronger gravitational force is between: {stronger_force}")

# Explanation of attraction
if force_earth_sun > force_moon_earth:
    explanation = "The Earth is more strongly attracted to the Sun than the Moon is to the Earth, due to the greater mass of the Sun compared to the Moon."
else:
    explanation = "The Moon is more strongly attracted to the Earth than the Earth is to the Sun, though the force comparison may be influenced by their relative masses and distances."

print(explanation)

```

```python

#4. Create a simple banking system using object-oriented concepts in Python. Design classes representing 
#   different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, 
#   and balance inquiry. Utilize inheritance to manage different account types efficiently.
#   Tasks
#   1. Define base class(es) for bank accounts with common attributes and methods
#   2. Implement subclasses for specific account types (e.g., SavingsAccount, CheckingAccount)
#   3. Provide methods for deposit, withdraw, and balance inquiry in each subclass
#   4. Test the banking system by creating instances of different account types and performing transactions.


class BankAccount:
    def __init__(self, account_number, account_holder, balance=0.0):
        self._account_number = account_number
        self._account_holder = account_holder
        self._balance = balance

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount:.2f}. New balance: {self._balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self._balance:
            self._balance -= amount
            print(f"Withdrew {amount:.2f}. New balance: {self._balance:.2f}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def inquiry_balance(self):
        return self._balance

    def display_info(self):
        print(f"Account Number: {self._account_number}")
        print(f"Account Holder: {self._account_holder}")
        print(f"Balance: {self._balance:.2f}")


class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, interest_rate=0.02):
        super().__init__(account_number, account_holder, balance)
        self._interest_rate = interest_rate

    def add_interest(self):
        interest = self._balance * self._interest_rate
        self.deposit(interest)
        print(f"Interest of {interest:.2f} added. New balance: {self._balance:.2f}")


class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder, balance=0.0, overdraft_limit=500.0):
        super().__init__(account_number, account_holder, balance)
        self._overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if 0 < amount <= self._balance + self._overdraft_limit:
            self._balance -= amount
            print(f"Withdrew {amount:.2f}. New balance: {self._balance:.2f}")
        else:
            print("Invalid withdrawal amount or exceeds overdraft limit.")


# Testing the banking system
def main():
    # Create a savings account
    savings = SavingsAccount("SA12345", "Sagar", 1000.0)
    savings.display_info()
    savings.deposit(500.0)
    savings.withdraw(200.0)
    savings.add_interest()
    savings.display_info()
    print()

    # Create a checking account
    checking = CheckingAccount("CA12345", "Amit", 500.0)
    checking.display_info()
    checking.deposit(300.0)
    checking.withdraw(700.0)
    checking.withdraw(200.0)
    checking.display_info()

if __name__ == "__main__":
    main()
```

```python
# 7. Write a code for  Hotel Management System using OOPS 
#   Create a Room class that has attributes such as room number, room type, rate, and availability (private)
#   Implement methods to book a room, check in a guest, and check out a guest
#   Use encapsulation to hide the room's unique identification number
#  Inherit from the Room class to create a SuiteRoom class and a StandardRoom class, each with their own specific attributes and methods.

class Room:
    __room_count = 0

    def __init__(self, room_number, room_type, rate):
        self.__id = Room.__room_count
        self.room_number = room_number
        self.room_type = room_type
        self.rate = rate
        self.__availability = True
        Room.__room_count += 1

    def get_id(self):
        return self.__id

    def is_available(self):
        return self.__availability

    def book_room(self):
        if self.__availability:
            self.__availability = False
            print(f"Room {self.room_number} has been booked.")
        else:
            print(f"Room {self.room_number} is already booked.")

    def check_in(self):
        if not self.__availability:
            print(f"Guest checked into room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is not booked yet.")

    def check_out(self):
        if not self.__availability:
            self.__availability = True
            print(f"Guest checked out of room {self.room_number}.")
        else:
            print(f"Room {self.room_number} is already available.")

    def display_room(self):
        availability = "Available" if self.__availability else "Booked"
        print(f"Room Number: {self.room_number}, Room Type: {self.room_type}, Rate: {self.rate}, Availability: {availability}")


class SuiteRoom(Room):
    def __init__(self, room_number, rate, has_living_area):
        super().__init__(room_number, "Suite", rate)
        self.has_living_area = has_living_area

    def display_room(self):
        super().display_room()
        print(f"Has Living Area: {self.has_living_area}")


class StandardRoom(Room):
    def __init__(self, room_number, rate, has_tv):
        super().__init__(room_number, "Standard", rate)
        self.has_tv = has_tv

    def display_room(self):
        super().display_room()
        print(f"Has TV: {self.has_tv}")


class Hotel:
    def __init__(self):
        self.rooms = []

    def add_room(self, room):
        self.rooms.append(room)
        print(f"Added: Room {room.room_number} ({room.room_type})")

    def book_room(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                room.book_room()
                return
        print("Room not found.")

    def check_in(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                room.check_in()
                return
        print("Room not found.")

    def check_out(self, room_number):
        for room in self.rooms:
            if room.room_number == room_number:
                room.check_out()
                return
        print("Room not found.")

    def display_rooms(self):
        print("\nHotel Rooms:")
        for room in self.rooms:
            room.display_room()
        print("\n")


# Testing the Hotel Management System
def main():
    hotel = Hotel()

    # Adding rooms to the hotel
    suite1 = SuiteRoom(101, 200, True)
    suite2 = SuiteRoom(102, 250, False)
    standard1 = StandardRoom(201, 100, True)
    standard2 = StandardRoom(202, 150, False)

    hotel.add_room(suite1)
    hotel.add_room(suite2)
    hotel.add_room(standard1)
    hotel.add_room(standard2)

    # Display rooms
    hotel.display_rooms()

    # Book a room
    hotel.book_room(101)
    hotel.book_room(201)

    # Check in a guest
    hotel.check_in(101)
    hotel.check_in(201)

    # Check out a guest
    hotel.check_out(101)
    hotel.check_out(201)

    # Display rooms after check in and check out
    hotel.display_rooms()


if __name__ == "__main__":
    main()
```

```python
# 8. Write a code for  Fitness Club Management System using OOPS
#    Create a Member class that has attributes such as name, age, membership type, and membership status (private)
#    Implement methods to register a new member, renew a membership, and cancel a membership
#    Use encapsulation to hide the member's unique identification number
#   Inherit from the Member class to create a FamilyMember class and an IndividualMember class, each with 
#   their own specific attributes and methods



class Member:
    __member_count = 0

    def __init__(self, name, age, membership_type):
        self.__id = Member.__member_count
        self.name = name
        self.age = age
        self.membership_type = membership_type
        self.__membership_status = "Active"
        Member.__member_count += 1

    def get_id(self):
        return self.__id

    def get_membership_status(self):
        return self.__membership_status

    def register_member(self):
        print(f"Member {self.name} has been registered with ID: {self.__id}.")

    def renew_membership(self):
        if self.__membership_status == "Active":
            print(f"Membership for {self.name} is already active.")
        else:
            self.__membership_status = "Active"
            print(f"Membership for {self.name} has been renewed.")

    def cancel_membership(self):
        if self.__membership_status == "Cancelled":
            print(f"Membership for {self.name} is already cancelled.")
        else:
            self.__membership_status = "Cancelled"
            print(f"Membership for {self.name} has been cancelled.")

    def display_member(self):
        print(f"ID: {self.__id}, Name: {self.name}, Age: {self.age}, Membership Type: {self.membership_type}, Status: {self.__membership_status}")


class FamilyMember(Member):
    def __init__(self, name, age, membership_type, family_size):
        super().__init__(name, age, membership_type)
        self.family_size = family_size

    def display_member(self):
        super().display_member()
        print(f"Family Size: {self.family_size}")


class IndividualMember(Member):
    def __init__(self, name, age, membership_type, personal_trainer):
        super().__init__(name, age, membership_type)
        self.personal_trainer = personal_trainer

    def display_member(self):
        super().display_member()
        print(f"Personal Trainer: {self.personal_trainer}")


class FitnessClub:
    def __init__(self):
        self.members = []

    def add_member(self, member):
        self.members.append(member)
        member.register_member()

    def renew_member(self, member_id):
        for member in self.members:
            if member.get_id() == member_id:
                member.renew_membership()
                return
        print("Member not found.")

    def cancel_member(self, member_id):
        for member in self.members:
            if member.get_id() == member_id:
                member.cancel_membership()
                return
        print("Member not found.")

    def display_members(self):
        print("\nFitness Club Members:")
        for member in self.members:
            member.display_member()
        print("\n")



def main():
    fitness_club = FitnessClub()

    # Adding members to the fitness club
    member1 = FamilyMember("Akhilesh yadav", 45, "Family", 4)
    member2 = IndividualMember("Mukesh Ambani", 30, "Individual", "Yes")
    member3 = IndividualMember("Narendra Modi", 25, "Individual", "No")

    fitness_club.add_member(member1)
    fitness_club.add_member(member2)
    fitness_club.add_member(member3)

    # Display members
    fitness_club.display_members()

    # Renew membership
    fitness_club.renew_member(1)

    # Cancel membership
    fitness_club.cancel_member(0)

    # Display members after renewal and cancellation
    fitness_club.display_members()


if __name__ == "__main__":
    main()
```

```python
# 9.Write a code for  Event Management System using OOPS
#   Create an Event class that has attributes such as name, date, time, location, and list of attendees (private)
#   Implement methods to create a new event, add or remove attendees, and get the total number of 
#   attendees
#   Use encapsulation to hide the event's unique identification number
#   Inherit from the Event class to create a PrivateEvent class and a PublicEvent class, each with their own 
#   specific attributes and methods.


class Event:
    __event_count = 0

    def __init__(self, name, date, time, location):
        self.__id = Event.__event_count
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []
        Event.__event_count += 1

    def get_id(self):
        return self.__id

    def create_event(self):
        print(f"Event '{self.name}' created with ID: {self.__id}")

    def add_attendee(self, attendee):
        self.__attendees.append(attendee)
        print(f"Attendee '{attendee}' added to event '{self.name}'")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"Attendee '{attendee}' removed from event '{self.name}'")
        else:
            print(f"Attendee '{attendee}' not found in event '{self.name}'")

    def get_total_attendees(self):
        return len(self.__attendees)

    def display_event(self):
        print(f"ID: {self.__id}, Name: {self.name}, Date: {self.date}, Time: {self.time}, Location: {self.location}, Attendees: {self.get_total_attendees()}")


class PrivateEvent(Event):
    def __init__(self, name, date, time, location, invite_only):
        super().__init__(name, date, time, location)
        self.invite_only = invite_only

    def display_event(self):
        super().display_event()
        print(f"Invite Only: {self.invite_only}")


class PublicEvent(Event):
    def __init__(self, name, date, time, location, max_capacity):
        super().__init__(name, date, time, location)
        self.max_capacity = max_capacity

    def display_event(self):
        super().display_event()
        print(f"Max Capacity: {self.max_capacity}")


class EventManager:
    def __init__(self):
        self.events = []

    def add_event(self, event):
        self.events.append(event)
        event.create_event()

    def add_attendee(self, event_id, attendee):
        for event in self.events:
            if event.get_id() == event_id:
                event.add_attendee(attendee)
                return
        print("Event not found.")

    def remove_attendee(self, event_id, attendee):
        for event in self.events:
            if event.get_id() == event_id:
                event.remove_attendee(attendee)
                return
        print("Event not found.")

    def display_events(self):
        print("\nEvent List:")
        for event in self.events:
            event.display_event()
        print("\n")


# Testing the Event Management System
def main():
    manager = EventManager()

    # Creating events
    event1 = PrivateEvent("Private Meeting", "2023-08-15", "10:00 AM", "Room 101", invite_only=True)
    event2 = PublicEvent("Open Seminar", "2023-08-20", "02:00 PM", "Auditorium", max_capacity=100)

    manager.add_event(event1)
    manager.add_event(event2)

    # Adding attendees
    manager.add_attendee(event1.get_id(), "John Doe")
    manager.add_attendee(event1.get_id(), "Jane Smith")
    manager.add_attendee(event2.get_id(), "Alice Johnson")
    manager.add_attendee(event2.get_id(), "Bob Brown")

    # Display events with attendees
    manager.display_events()

    # Removing an attendee
    manager.remove_attendee(event1.get_id(), "John Doe")

    # Display events after removal
    manager.display_events()


if __name__ == "__main__":
    main()

```

```python
# 10. Write a code for Airline Reservation System using OOPS
#     Create a Flight class that has attributes such as flight number, departure and arrival airports, departure and 
#     arrival times, and available seats (private)
#     Implement methods to book a seat, cancel a reservation, and get the remaining available seats
#     Use encapsulation to hide the flight's unique identification number
#     Inherit from the Flight class to create a DomesticFlight class and an InternationalFlight class, each with their own specific attributes and methods.


class Flight:
    __flight_count = 0

    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats):
        self.__id = Flight.__flight_count
        self.flight_number = flight_number
        self.departure_airport = departure_airport
        self.arrival_airport = arrival_airport
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.__available_seats = available_seats
        Flight.__flight_count += 1

    def get_id(self):
        return self.__id

    def book_seat(self):
        if self.__available_seats > 0:
            self.__available_seats -= 1
            print(f"Seat booked. Remaining seats: {self.__available_seats}")
        else:
            print("No available seats to book.")

    def cancel_reservation(self):
        self.__available_seats += 1
        print(f"Reservation cancelled. Remaining seats: {self.__available_seats}")

    def get_remaining_seats(self):
        return self.__available_seats

    def display_flight(self):
        print(f"ID: {self.__id}, Flight Number: {self.flight_number}, Departure: {self.departure_airport} at {self.departure_time}, Arrival: {self.arrival_airport} at {self.arrival_time}, Available Seats: {self.__available_seats}")


class DomesticFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, domestic_tax):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats)
        self.domestic_tax = domestic_tax

    def display_flight(self):
        super().display_flight()
        print(f"Domestic Tax: {self.domestic_tax}")


class InternationalFlight(Flight):
    def __init__(self, flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats, visa_required):
        super().__init__(flight_number, departure_airport, arrival_airport, departure_time, arrival_time, available_seats)
        self.visa_required = visa_required

    def display_flight(self):
        super().display_flight()
        print(f"Visa Required: {self.visa_required}")


class AirlineReservationSystem:
    def __init__(self):
        self.flights = []

    def add_flight(self, flight):
        self.flights.append(flight)
        print(f"Flight {flight.flight_number} added.")

    def book_flight(self, flight_id):
        for flight in self.flights:
            if flight.get_id() == flight_id:
                flight.book_seat()
                return
        print("Flight not found.")

    def cancel_flight(self, flight_id):
        for flight in self.flights:
            if flight.get_id() == flight_id:
                flight.cancel_reservation()
                return
        print("Flight not found.")

    def display_flights(self):
        print("\nAvailable Flights:")
        for flight in self.flights:
            flight.display_flight()
        print("\n")


# Testing the Airline Reservation System
def main():
    system = AirlineReservationSystem()

    # Creating flights
    domestic_flight = DomesticFlight("D123", "New York", "Bengaluru", "09:00", "11:00", 50, 50.0)
    international_flight = InternationalFlight("I456", "Nagpur", "Tokyo", "14:00", "18:00", 30, True)

    system.add_flight(domestic_flight)
    system.add_flight(international_flight)

    # Display flights
    system.display_flights()

    # Book a seat
    system.book_flight(domestic_flight.get_id())

    # Cancel a reservation
    system.cancel_flight(international_flight.get_id())

    # Display flights after booking and cancellation
    system.display_flights()


if __name__ == "__main__":
    main()


```
